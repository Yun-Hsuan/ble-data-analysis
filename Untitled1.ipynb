```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import movingpandas as mpd
import matplotlib.pyplot as plt

# Step 1: Generate sample terminal locations
num_terminals = 10
terminals = {
    f"terminal_{i+1}": (np.random.uniform(0, 100), np.random.uniform(0, 100))
    for i in range(num_terminals)
}

# Step 2: Generate sample BLE data for 100 memberIds over time
num_members = 100
timestamps = pd.date_range("2024-12-04 11:00", periods=12, freq="1H")
ble_data = []

for member_id in range(1, num_members + 1):
    for t in timestamps:
        terminal_id = np.random.choice(list(terminals.keys()))
        x, y = terminals[terminal_id]
        rssi = np.random.uniform(-80, -60)  # Simulate RSSI values
        ble_data.append([member_id, terminal_id, t, x + np.random.uniform(-1, 1), y + np.random.uniform(-1, 1), rssi])

df = pd.DataFrame(ble_data, columns=["memberId", "terminalId", "eventTime", "x", "y", "rssi"])

# Step 3: Convert to GeoDataFrame
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.x, df.y), crs="EPSG:4326")

# Step 4: Create TrajectoryCollection
tc = mpd.TrajectoryCollection(gdf, "memberId", t="eventTime")

# Step 5: Plot generalized aggregated trajectories
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
tc.plot(ax=ax, column="memberId", legend=True, cmap="tab20", linewidth=2, alpha=0.7)
ax.set_title("Generalized aggregated trajectories (Simulated)")
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from collections import Counter
from shapely.geometry import Point, LineString

# Example: Simulate aggregated trajectories with weights
edges = []

for member_id, group in gdf.groupby('memberId'):
    points = group.sort_values(by='eventTime')['geometry'].tolist()
    for i in range(len(points) - 1):
        edges.append((points[i], points[i + 1]))



# Count occurrences of each edge (trajectory segment)
edge_counts = Counter(edges)

# Plot aggregated trajectories with weights
fig, ax = plt.subplots(figsize=(10, 8))

for edge, count in edge_counts.items():
    line = LineString(edge)
    ax.plot(*line.xy, linewidth=count / 5, alpha=0.6)  # Adjust linewidth for visibility

# Add point markers
gdf.plot(ax=ax, color='red', markersize=20, alpha=0.7)

ax.set_title("Generalized aggregated trajectories (Weighted)")
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import movingpandas as mpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
from datetime import datetime, timedelta
import random

# Step 1: Generate sample terminal coordinates
num_terminals = 10
terminal_coords = {f"Terminal_{i+1}": (random.uniform(0, 100), random.uniform(0, 100)) for i in range(num_terminals)}

# Step 2: Generate sample trajectories (100 unique members)
num_members = 100
data = []
for member_id in range(1, num_members + 1):
    start_terminal = random.choice(list(terminal_coords.keys()))
    end_terminal = random.choice(list(terminal_coords.keys()))
    start_x, start_y = terminal_coords[start_terminal]
    end_x, end_y = terminal_coords[end_terminal]
    
    # Create a simple trajectory with two points (start and end terminals)
    data.append({
        "memberId": member_id,
        "trajectory_id": f"Traj_{member_id}",
        "geometry": LineString([(start_x, start_y), (end_x, end_y)]),
        "start_time": datetime(2024, 12, 4, random.randint(11, 21), random.randint(0, 59)),
        "end_time": datetime(2024, 12, 4, random.randint(11, 22), random.randint(0, 59))
    })

# Step 3: Create GeoDataFrame
gdf = gpd.GeoDataFrame(data, crs="EPSG:3857")  # Using a planar CRS for x, y coordinates

# Step 4: Create TrajectoryCollection
tc = mpd.TrajectoryCollection(gdf, traj_id_col="trajectory_id", t="start_time")

# Step 5: Generalize trajectories using MinDistanceGeneralizer
generalizer = mpd.MinDistanceGeneralizer(tc)
generalized = generalizer.generalize(tolerance=10)

# Step 6: Aggregate trajectories using TrajectoryCollectionAggregator
aggregator = mpd.TrajectoryCollectionAggregator(
    generalized,
    max_distance=20,
    min_distance=5,
    min_stop_duration=timedelta(minutes=1)
)

# Step 7: Get flows and clusters for visualization
flows = aggregator.get_flows_gdf()
clusters = aggregator.get_clusters_gdf()

# Step 8: Plot the aggregated trajectories
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
flows.plot(ax=ax, linewidth=1, color='blue', alpha=0.7, label='Aggregated Flows')
clusters.plot(ax=ax, color='red', markersize=100, label='Cluster Centroids')
for terminal, (x, y) in terminal_coords.items():
    ax.scatter(x, y, color='black')
    ax.text(x, y, terminal, fontsize=8, ha='right')
ax.set_title("Aggregated Trajectories (Simulated)")
ax.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from shapely.geometry import Point, LineString
import random

# Step 1: Simulate terminal coordinates
num_terminals = 10
terminal_coords = {
    f"Terminal_{i+1}": (random.randint(0, 100), random.randint(0, 100))
    for i in range(num_terminals)
}
terminals_df = pd.DataFrame(
    terminal_coords.items(), columns=["terminal_id", "coords"]
)
terminals_df[["x", "y"]] = terminals_df["coords"].apply(pd.Series)

# Step 2: Simulate human flow data
num_members = 100
flow_data = []
for member_id in range(1, num_members + 1):
    num_movements = random.randint(2, 5)
    visited_terminals = random.sample(list(terminal_coords.keys()), num_movements)
    for i in range(len(visited_terminals) - 1):
        flow_data.append(
            {
                "member_id": member_id,
                "from_terminal": visited_terminals[i],
                "to_terminal": visited_terminals[i + 1],
            }
        )

flow_df = pd.DataFrame(flow_data)

# Step 3: Count flows between terminals
flow_counts = flow_df.groupby(["from_terminal", "to_terminal"]).size().reset_index(name="count")

# Step 4: Plot aggregated human flows
fig, ax = plt.subplots(figsize=(10, 8))

# Plot terminals
for terminal, (x, y) in terminal_coords.items():
    ax.scatter(x, y, color="red")
    ax.text(x, y + 2, terminal, ha="center")

# Plot flows
for _, row in flow_counts.iterrows():
    from_terminal = row["from_terminal"]
    to_terminal = row["to_terminal"]
    count = row["count"]
    
    x1, y1 = terminal_coords[from_terminal]
    x2, y2 = terminal_coords[to_terminal]
    
    line = LineString([(x1, y1), (x2, y2)])
    gpd.GeoSeries([line]).plot(ax=ax, linewidth=count / 2, alpha=0.7, color="blue")

ax.set_title("Aggregated Human Flows between Terminals")
ax.set_xlabel("X Coordinate")
ax.set_ylabel("Y Coordinate")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import networkx as nx
import random

# Step 1: Generate sample terminal coordinates
num_terminals = 10
terminal_coords = {f"Terminal_{i}": (random.randint(0, 100), random.randint(0, 100)) for i in range(num_terminals)}

# Step 2: Simulate human flows between terminals
terminal_pairs = [(f"Terminal_{i}", f"Terminal_{j}") for i in range(num_terminals) for j in range(i + 1, num_terminals)]
flows = {pair: random.randint(1, 100) for pair in terminal_pairs}  # Random flow values

# Step 3: Create a graph and add edges with flow as weight
G = nx.Graph()
for terminal, coord in terminal_coords.items():
    G.add_node(terminal, pos=coord)

for (t1, t2), flow in flows.items():
    G.add_edge(t1, t2, weight=flow)

# Step 4: Normalize flow values for colormap
flow_values = [G[u][v]['weight'] for u, v in G.edges()]
norm = mcolors.Normalize(vmin=min(flow_values), vmax=max(flow_values))
cmap = plt.cm.Reds  # Using Reds colormap (blue to red)

# Step 5: Plot the graph with color-coded edges
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(10, 8))
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='red')
nx.draw_networkx_labels(G, pos)

for (u, v, weight) in G.edges(data='weight'):
    nx.draw_networkx_edges(
        G, pos, edgelist=[(u, v)],
        width=2, edge_color=[cmap(norm(weight))], alpha=0.8
    )

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, label='Flow Intensity')

plt.title("Aggregated Trajectories with Heatmap")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import networkx as nx
import seaborn as sns
import random

# Step 1: Generate sample terminal coordinates
num_terminals = 10
terminal_coords = {f"Terminal_{i}": (random.randint(0, 100), random.randint(0, 100)) for i in range(num_terminals)}

# Step 2: Simulate human flows between terminals
terminal_pairs = [(f"Terminal_{i}", f"Terminal_{j}") for i in range(num_terminals) for j in range(i + 1, num_terminals)]
flows = {pair: random.randint(1, 100) for pair in terminal_pairs}  # Random flow values

# Step 3: Calculate total flow for each terminal (for node size)
terminal_flow = {terminal: 0 for terminal in terminal_coords}
for (t1, t2), flow in flows.items():
    terminal_flow[t1] += flow
    terminal_flow[t2] += flow

# Step 4: Create a graph and add edges with flow as weight
G = nx.Graph()
for terminal, coord in terminal_coords.items():
    G.add_node(terminal, pos=coord, total_flow=terminal_flow[terminal])

for (t1, t2), flow in flows.items():
    G.add_edge(t1, t2, weight=flow)

# Step 5: Normalize flow values for colormap and sizes
flow_values = [G[u][v]['weight'] for u, v in G.edges()]
node_sizes = [terminal_flow[node] * 2 for node in G.nodes()]  # Scale node sizes
edge_widths = [weight / 10 for weight in flow_values]  # Scale edge widths

norm = mcolors.Normalize(vmin=min(flow_values), vmax=max(flow_values))
cmap = sns.cubehelix_palette(as_cmap=True)  # Seaborn colormap

# Step 6: Plot the graph with color-coded edges and variable node sizes
pos = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(12, 10))

# Draw nodes with varying sizes
nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color='red', alpha=0.8)

# Draw labels
nx.draw_networkx_labels(G, pos)

# Draw edges with varying widths and colors
for (u, v, weight) in G.edges(data='weight'):
    nx.draw_networkx_edges(
        G, pos, edgelist=[(u, v)],
        width=edge_widths.pop(0), edge_color=[cmap(norm(weight))], alpha=0.6
    )

# Add colorbar
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
plt.colorbar(sm, label='Flow Intensity')

plt.title("Aggregated Trajectories with Heatmap and Variable Node Sizes")
plt.xlabel("X Coordinate")
plt.ylabel("Y Coordinate")
plt.show()
```
