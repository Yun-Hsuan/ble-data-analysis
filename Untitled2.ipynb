```python
import pandas as pd
from pathlib import Path
from openpyxl.utils import get_column_letter
from openpyxl.styles import NamedStyle, numbers
import pandas as pd
from datetime import datetime, timedelta
from src.data_processing.cleaners.ble_cleaner import BLECleaner
from src.data_processing.cleaners.transaction_cleaner import TransactionCleaner
from src.business.tenant_indicators.visit_rate import VisitRateIndicator
from src.business.tenant_indicators.pass_by import PassByIndicator
from src.business.tenant_indicators.dwell_rate import DwellRateIndicator
from src.business.tenant_indicators.bagging_rate import BaggingRateIndicator
from src.report.report_manager import ReportManager
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import networkx as nx
import seaborn as sns
import random
from PIL import Image
import pandas as pd
from matplotlib.colors import LogNorm  # LogNorm for logarithmic normalization
```

```python
def date_range(start_date: str, end_date: str):
    """
    Generate a range of dates between start_date and end_date (inclusive).
    :param start_date: Start date in the format 'YYYY-MM-DD'.
    :param end_date: End date in the format 'YYYY-MM-DD'.
    :return: A list of date strings in the format 'YYYY-MM-DD'.
    """
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")
    return [(start + timedelta(days=i)).strftime("%Y-%m-%d") for i in range((end - start).days + 1)]
```

```python
date_list = date_range("2024-12-22", "2024-12-22")
start_time = pd.to_datetime("2024-12-22 11:00:00")
end_time = pd.to_datetime("2024-12-22 22:00:00")
file_path = "./data/processed/tenant_info/tenant_rssi_thresholds.xlsx"
threshold_data = pd.read_excel(file_path)
pass_by_rssi_threshold = threshold_data.set_index('terminalId')['pass_by_rssi_threshold'].to_dict()
for date in date_list:

    # Step 1: Load floor map image
    floor_map = Image.open("data/B1_floor_map2.png")

    # Step 2: Load terminal coordinates from Excel
    excel_path = "./data/b1_floor_map_cordinate.xlsx"
    df_coords = pd.read_excel(excel_path)

    # Step 2: Remove invalid entries where x or y is '-'
    df_coords_cleaned = df_coords[(df_coords['x'] != '-') & (df_coords['y'] != '-')]
    
    # Step 3: Convert terminal coordinates to a dictionary {terminalId: (x, y)}
    terminal_coords = {int(row["terminalId"]): (row["x"], row["y"]) for _, row in df_coords_cleaned.iterrows()}
    
    print(f"Processing date: {date}")

    ble_cleaner = BLECleaner(directory=f"data/processed/ble")
    print("Loading ble data...")
    ble_cleaner.load_data(pattern=fr"{date}.*\.csv")
    ble_data = ble_cleaner.get_data()
    original_member_count = ble_data['memberId'].nunique()
    
    excluded_terminals = [201246, 201258, 201242]
    ble_data = ble_data[~ble_data['terminalId'].isin(excluded_terminals)]
    
    ble_data.loc[:, 'eventTime'] = pd.to_datetime(ble_data['eventTime'], errors='coerce')
    ble_data = ble_data.dropna(subset=['eventTime'])
    ble_data['eventTime'] = pd.to_datetime(ble_data['eventTime'])
    if True and 'eventTime' in ble_data.columns:
        ble_data = ble_data[(ble_data['eventTime'] >= start_time) & (ble_data['eventTime'] <= end_time)]

    
    filtered_by_rssi = ble_data[
        ble_data.apply(lambda row: row['rssi'] > pass_by_rssi_threshold.get(row['terminalId'], -75), axis=1)
    ]
    
    # Step 2: Group by 'memberId' and filter out groups with only one unique 'terminalId'
    
    filtered_data = filtered_by_rssi[
        filtered_by_rssi['accessAddress'].str[14:20] == 'ff4c00'
    ]

    filtered_data = (
        filtered_data.groupby('memberId')
        .filter(lambda group: group['terminalId'].nunique() > 1)
    )
    
    filtered_member_count = filtered_data['memberId'].nunique()

    print(filtered_data)
    # Step 5: Print result summary
    print(f"Original unique member count: {original_member_count}")
    print(f"Filtered unique member count: {filtered_member_count}")
    print(f"Removed {original_member_count - filtered_member_count} unique members.")

    grouped = filtered_data.groupby('memberId').apply(lambda x: x.sort_values(by='eventTime'))

    # Step 6: Generate flow pairs
    flow_pairs = []
    for member_id, group in grouped.groupby(level=0):
        terminals = group['terminalId'].tolist()
        for i in range(len(terminals) - 1):
            flow_pairs.append((terminals[i], terminals[i + 1]))

    # Step 7: Count flow intensity for each pair
    flows = pd.Series(flow_pairs).value_counts().to_dict()

    # Step 8: Calculate total flow for each terminal (for node size)
    print(f"Processing Calculation: {date}")

    terminal_flow = {terminal: 0 for terminal in terminal_coords}
    for (t1, t2), flow in flows.items():
        terminal_flow[t1] += flow
        terminal_flow[t2] += flow

    # Step 5: Create graph with positions
    G = nx.Graph()
    for terminal, coords in terminal_coords.items():
        G.add_node(terminal, pos=coords)
    
    for (t1, t2), flow in flows.items():
        G.add_edge(t1, t2, weight=flow)
    
    # Step 6: Normalize flow values for colormap, node size, and edge width
    flow_values = [G[u][v]['weight'] for u, v in G.edges()]
    min_size, max_size = 200, 800
    scaling_factor = 0.02  
    node_sizes = [
        (min_size + (terminal_flow[node] / max(terminal_flow.values())) * (max_size - min_size)) * scaling_factor
        for node in G.nodes()
    ]
    edge_widths = [0.5 + (weight / max(flow_values)) * 15 for weight in flow_values]  # Scale edge widths
    #norm = mcolors.Normalize(vmin=min(flow_values), vmax=max(flow_values))
    #cmap = sns.cubehelix_palette(as_cmap=True)  # Seaborn colormap
    #cmap = plt.cm.get_cmap("RdBu")
    # Use logarithmic normalization for better contrast
    norm = LogNorm(vmin=max(min(flow_values), 1), vmax=max(flow_values))  # Ensure vmin > 0
    #cmap = sns.cubehelix_palette(start=0.5, rot=-0.75, gamma=0.8, as_cmap=True)  # Adjust colormap
    cmap = sns.cubehelix_palette(start=0.0, rot=1.0, gamma=0.8, as_cmap=True)
    # Step 7: Plot the graph on the floor map
    fig, ax = plt.subplots(figsize=(12, 5))
    
    # Plot the floor map as background
    ax.imshow(floor_map, extent=[0, 36, 0, 16])  # Set extent to match the coordinate range
    
    # Draw nodes with varying sizes
    nx.draw_networkx_nodes(G, pos=nx.get_node_attributes(G, 'pos'), node_size=node_sizes,
                           node_color='red', alpha=0.8, ax=ax)
    
    # Draw labels
    #nx.draw_networkx_labels(G, pos=nx.get_node_attributes(G, 'pos'), ax=ax)
    
    # Draw edges with varying widths and colors
    for (u, v, weight), width in zip(G.edges(data='weight'), edge_widths):
        nx.draw_networkx_edges(
            G, pos=nx.get_node_attributes(G, 'pos'), edgelist=[(u, v)],
            width=width, edge_color=[cmap(norm(weight))], alpha=0.6, ax=ax
        )
    
    # Add colorbar
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    sm.set_array([])
    plt.colorbar(sm, ax=ax, label='Flow Intensity')
    
    plt.title(f"Shopping Aggregated Trajectories with Heatmap {date}")
    #plt.xlabel("X Coordinate")
    #plt.ylabel("Y Coordinate")
    plt.show()
```
